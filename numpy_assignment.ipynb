```python
import numpy as np
```

```python
expenses = np.array([2000, 5000, 10000, 1000, 500])

#find the average of expenses
average = np.mean(expenses)
print(f" The average monthly expenses: {average}")

#find the maximum number of expenses
maximum = np.max(expenses)
print(f" The maximum amount in monthly expenses: {maximum}")

#find the minimum number of expenses
minimum =np.min(expenses)
print(f" The minimum amount in monthly expenses: {minimum}")

#find the total sum of expenses
total_sum =np.sum(expenses)
print(f" The total sum in monthly expenses : {total_sum}")

# find the first three monthly expenses 
# using slicing
print(f" The first three monthly expenses are {expenses[0:3]}")
```
