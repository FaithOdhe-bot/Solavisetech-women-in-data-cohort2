```python
import pandas as pd
import numpy as np
data = {
'Order_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
'Customer_Name': ['Amara', 'Bola', None, 'Diana', 'Efua', 'Fatima', 'Grace',
None, 'Ifeoma', 'Joyce'],
'Country': ['Ghana', 'Nigeria', 'Kenya', None, 'Ghana', 'Cameroon', None,
'Kenya', 'Nigeria', 'Ghana'],
'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', None, 'Phone', 'Tablet',
'Laptop', None, 'Phone'],
'Quantity': [2, 1, None, 3, 2, None, 1, 4, 2, None],
'Price': [1500, 800, 600, 1500, None, 800, 600, 1500, None, 800],
'Delivery_Days': [3, None, 7, 5, 3, None, 6, None, 4, 5]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
```

i try to use df['Customer_Name'].fillna('unknown', inplace= True ) to apply to a single column selected but with a newer versions of pandas doing this modifies a copy of the data rather than the DataFrame

```python
quantity_median = df['Quantity'].median
df['Quantity'] = df['Quantity'].fillna(quantity_median)
```

the trends was that the quantities usually stay between 1 to 4 items

```python
price_median =df['Price'].median
df['Price'] = df['Price'].fillna(price_median)
```

it was noticed the range of the object was 600-1500 and the median gives the actual middle price 

```python
delivery_mean = df['Delivery_Days'].mean()
df['Delivery_Days'] = df['Delivery_Days'].fillna(delivery_mean)
```

the delivery times all sit tightly between 3 and 7 days, checking there no extereme outliners so a mean was a great choice 

```python
country_mode = df['Country'].mode()[0]
df['Country'] = df['Country'].fillna(country_mode)
```

i chose to fill the missing countries with the most frquent value and dropping the rows will cause a lose of too much data

```python
df['Product'] = df['Product'].fillna('Unknown')
```

Using a placeholder than making a false assumption is a justification
