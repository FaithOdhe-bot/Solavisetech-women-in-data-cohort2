```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Order_ID':['ORD001','ORD002','ORD003','ORD004','ORD005','ORD006','ORD007','ORD008','ORD009','ORD010'],
    'Product':['Laptop','Rice Bag','Headphones','Notebook','Blender', 'Rice Bag','Laptop','Notebook','Blender','Headphones'],
    'Category':['Electronics','Food','Electronics','Stationery','Appliances','Food','Electronics','Stationery','Appliances','Electronics'],
    'Region':['North','South','East','West','North','East','South','North','West','North'],
    'Units_Sold': [5, 30, 12, 50, 8, 20, 3, 80, 6, 15],
    'Unit_Price': [850.00, 12.50, 45.00, 3.20, 65.00, 12.50, 850.00, 3.20, 65.00, 45.00],
    'Month':['January','January','February','February','March','March','April','April','May','May']
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

# 1. Group, sum, and sort the data
region_sales = df.groupby('Region')['Units_Sold'].sum().sort_values(ascending=False)

# 2. Plot directly from the pandas Series
region_sales.plot(kind='bar', color='red', figsize=(8, 5))

# 3. Customize the labels
plt.title("Sales Performance by Region")
plt.xlabel("Region")
plt.ylabel("Total Units Sold")
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

```python
monthly_data = df.groupby('Month')['Unit_Price'].mean().reset_index()
#sort chronologically so the line connects correctly from January to May
month_order = ['January', 'February', 'March', 'April', 'May']
monthly_data['Month'] = pd.Categorical(monthly_data['Month'], categories=month_order, ordered=True)
monthly_data = monthly_data.sort_values('Month')
```

```python
plt.figure(figsize=(8, 5))
plt.plot(monthly_data['Month'], monthly_data['Unit_Price'], marker="o", label="Avg Unit Price", color="steelblue")

# 3. Customizing labels and titles
plt.title("Average Unit Price Over Months")
plt.xlabel("Month")
plt.ylabel("Unit Price (USD)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
category_units = df.groupby('Category')['Units_Sold'].sum().sort_values(ascending=False)


plt.figure(figsize=(6, 6))
plt.pie(
    category_units, 
    labels=category_units.index, 
    autopct='%1.1f%%',       
    startangle=140,         
    colors=['Red', 'Green', 'Blue', 'Yellow'] 
)

# 4. Customizing the chart
plt.title("Proportion of Total Units Sold by Category", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(
    df['Units_Sold'], 
    bins=5,               
    color='teal',        
    edgecolor='black'     
)


plt.title("Distribution of Units Sold", fontsize=14, weight='bold')
plt.xlabel("Units Sold (Ranges)")
plt.ylabel("Frequency (Number of Orders)")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Unit_Price'], df['Units_Sold'], color='purple', alpha=0.7, s=100)

# 2. Customize labels and grid lines
plt.title("Unit Price vs. Units Sold")
plt.xlabel("Unit Price (USD)")
plt.ylabel("Units Sold")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
```
